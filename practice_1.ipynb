```python
my_stat = my_stat.rename(columns={
    'V1': 'session_value',
    'V2': 'group',
    'V3': 'time',
    'V4': 'n_users'
})
```

```python
my_stat.session_value = my_stat.session_value.fillna(0)
```

```python
med = my_stat[my_stat['n_users'] > 0].n_users.median()
my_stat['n_users'] = np.where(my_stat['n_users'] < 0, med, my_stat['n_users'])
```
